```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
movies=movies.merge(credits,on='title')
```

```python
movies.head()
```

```python
#important colums to keep:
# genres
# id
# keywords
# title
# overview
# cdst
# crew
movies=movies[['genres','id','keywords','title','overview','cast','crew']]
```

```python
movies.head()
```

```python
#check for the missing values
movies.isnull().sum()
```

```python
#droping the missing colums
movies.dropna(inplace=True)
```

```python
#check for duplicate
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
#'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
#['Action,Adventure,Fantacy,scifi']
```

```python
#ast is library which convert strings into list by useing ast.literal_eval
import ast
def change(obj):
    temp=[]
    for i in ast.literal_eval(obj):
        temp.append(i['name'])
    return temp
```

```python
#applying the helper function to genres
movies['genres']=movies['genres'].apply(change)
```

```python
#applying the helper function to keywords
movies['keywords']=movies['keywords'].apply(change)
```

```python
#making a function for taking top 3 cast member of the flim
def helper_cast(obj):
    temp=[]
    l=ast.literal_eval(obj)
    for i in range(min(3, len(l))):
        temp.append(l[i]['name'])
    return temp
```

```python
movies['cast']=movies['cast'].apply(helper_cast)
```

```python
movies.head()
```

```python
# helper to extract director
def helper_crew(obj):
    temp=[]
    l=ast.literal_eval(obj)
    for i in l:
        if(i['job']=='Director'):
            temp.append(i['name'])
    return temp
    
```

```python
movies['crew']=movies['crew'].apply(helper_crew)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#removing the spaces
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
newdf=movies[['id','title','tags']]
newdf.head()
```

```python
newdf['tags']=newdf['tags'].apply(lambda x:" ".join(x))
```

# Text Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer  
cv=CountVectorizer(max_features=5000,stop_words='english')
vectors=cv.fit_transform(newdf['tags']).toarray()
change above code wth bert word to token 
```

```python
#Applying the stemming
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
newdf['tags']=newdf['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index=newdf[newdf['title']==movie].index[0]
    distance=similarity[movie_index]
    movies_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(newdf.iloc[i[0]].title)
        
```
